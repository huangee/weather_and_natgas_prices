```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import statsmodels.formula.api as smf
```

```python
df_Henry = pd.read_csv('./data/henry.csv', usecols=["Date","Price","Log_Price","Low_Temp","Hi_Temp","Precip_Type","Snow_Amt"])
df_TranscoZ3 = pd.read_csv('./data/transcoz3.csv', usecols=["Date","Price","Log_Price","Low_Temp","Hi_Temp","Precip_Type","Snow_Amt"])
df_TranscoZ6 = pd.read_csv('./data/transcoz6.csv', usecols=["Date","Price","Log_Price","Low_Temp","Hi_Temp","Precip_Type","Snow_Amt"])
df_PEPL = pd.read_csv('./data/pepl.csv', usecols=["Date","Price","Log_Price","Low_Temp","Hi_Temp","Precip_Type","Snow_Amt"])
df_Opal = pd.read_csv('./data/opal.csv', usecols=["Date","Price","Log_Price","Low_Temp","Hi_Temp","Precip_Type","Snow_Amt"])
```

### Industry has different rules and trends for summer (Apr 1 - Oct 31) versus winter (Nov 1 - Mar 31)

```python
## testing correlation split between summer (Apr-Oct) and winter (Nov-Mar) months

df_Henry['Date'] = pd.to_datetime(df_Henry['Date'])

df_Henry_winter = df_Henry[(df_Henry.Date.dt.month > 10) | (df_Henry.Date.dt.month < 4)]

df_Henry_winter.sample(10)
```

```python
sns.pairplot(df_Henry_winter);
```

```python
lm_henry_winter = smf.ols('Log_Price ~ Low_Temp', data=df_Henry_winter)

fit_henry_winter = lm_henry_winter.fit()

fit_henry_winter.summary()
```

```python
df_Henry_winter['Low_Temp_log'] = np.log(df_Henry_winter.Low_Temp)
```

```python
lm_henry_winter = smf.ols('Log_Price ~ Low_Temp_log', data=df_Henry_winter)

fit_henry_winter = lm_henry_winter.fit()

fit_henry_winter.summary()
```

```python
df_TranscoZ3['Date'] = pd.to_datetime(df_TranscoZ3['Date'])
df_TranscoZ3_winter = df_TranscoZ3[(df_TranscoZ3.Date.dt.month > 10) | (df_TranscoZ3.Date.dt.month < 4)]

df_PEPL['Date'] = pd.to_datetime(df_PEPL['Date'])
df_PEPL_winter = df_PEPL[(df_PEPL.Date.dt.month > 10) | (df_PEPL.Date.dt.month < 4)]

df_TranscoZ3_winter['Low_Temp_log'] = np.log(df_TranscoZ3_winter.Low_Temp)
df_PEPL_winter['Low_Temp_log'] = np.log(df_PEPL_winter.Low_Temp)
```

```python
lm_pepl_winter = smf.ols('Log_Price ~ Low_Temp_log', data=df_PEPL_winter)

fit_pepl_winter = lm_pepl_winter.fit()

fit_pepl_winter.summary()
```

```python
df_Henry_winter.to_csv('./data/henry_winter.csv', index=False)
df_TranscoZ3_winter.to_csv('./data/transcoz3_winter.csv', index=False)
df_PEPL_winter.to_csv('./data/pepl_winter.csv', index=False)
```
